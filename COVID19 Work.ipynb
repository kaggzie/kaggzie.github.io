```python
#Coerce object data into numeric form 

dfTHA['Phone number'] = pd.to_numeric(dfTHA['Phone number'], errors='coerce')
```

```python
#drop all rows with NAN data 
dfTHA = dfTHA.dropna(subset=['Phone number'])
dfTHA
```

```python
#Convert phone number from float into int

dfTHA['Phone number'] = dfTHA['Phone number'].astype(int)
dfTHA
```

```python
#Do a left join merge using Phone number as the join ID. 
#Why not an inner join?

merged_left = pd.merge(left=dfSMS, right=dfTHA, how='left', left_on='Phone Number', right_on='Phone number')
merged_left
```
